```python
import re
import pandas as pd
```

```python
f=open('gehu_wchat.txt', 'r' ,encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
#Regular Expression String stored in pattern
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages=re.split(pattern, data)[1:]
messages
```

```python
dates=re.findall(pattern, data)
dates
```

```python
#create new dataframe which inclue user-message and dates
df=pd.DataFrame({'user_message':messages, 'message_date': dates})
# df['message_date']=pd.to_datetime(df['message_date'], errors='coerce', format='%m/%d/%Y, %H:%M - ')
df['message_date']=pd.to_datetime(df['message_date'], errors='coerce').dt.date
df.rename(columns={'message_date':'date'}, inplace=True)
df
```

```python
df.shape
```

```python
#separate users and messages
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split('([\w\w]+?):\s', message)
    if entry[1:]:  #username
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_noticification')
        messages.append(entry[0])

df['user']=users
df['message']=messages
df.drop(columns=['user_message'], inplace=True)
df.head()
```

```python
#extract year
df['year']=df['date'].dt.year
```
